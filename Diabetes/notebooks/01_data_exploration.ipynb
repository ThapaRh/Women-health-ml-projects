```python
import pandas as pd

# read SAS XPORT files using pandas.read_sas with format='xport'
demo = pd.read_sas("../data/raw/DEMO_L.XPT", format="xport")
hba1c = pd.read_sas("../data/raw/GHB_L.XPT", format="xport")
diabetes = pd.read_sas("../data/raw/DIQ_L.XPT", format="xport")

demo.head()
```

```python
women = demo[demo['RIAGENDR']==2]
```

```python
# Merge demo and hba1c on SEQN
merged = pd.merge(women, hba1c, on='SEQN', how='inner')
# Merge the result with diabetes on SEQN
merged = pd.merge(merged, diabetes, on='SEQN', how='inner')
# Display the first few rows of the merged dataset
merged.head()
```

```python
# Define diabetes risk label
# HbA1c column is usually named 'LBXGH' or similar in NHANES GHB_L.XPT
# Self-reported diabetes is usually in DIQ010 (1 = Yes)

# Adjust column names if needed based on your dataset
merged['diabetes_risk'] = ((merged['LBXGH'] >= 6.5) | (merged['DIQ010'] == 1)).astype(int)

# Show the first few rows with the new label
merged[['SEQN', 'LBXGH', 'DIQ010', 'diabetes_risk']].head()
```
